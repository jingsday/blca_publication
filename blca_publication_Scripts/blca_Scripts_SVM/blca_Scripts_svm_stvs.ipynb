```python
import os 
import pandas as pd
```

```python
datadir = '/home/jing/Phd_project/project_UCD_blca/blca_publication_OUTPUT/blca_publication_OUTPUT_svm/'
os.chdir(datadir)
```

```python
stv_df = pd.DataFrame(columns=['blca_invasiveness','blca_oncogenesis','blca_survival','Genes'],
                      )
```

```python
gene_info_df=pd.read_csv('/home/jing/Phd_project/project_GBM/gbm_DATA/gbm_DATA_LINCS/2020/info/geneinfo_beta.txt', sep = "\t", index_col = 1)

gene_info_df = gene_info_df.sort_values(by = ["gene_symbol"])
display(gene_info_df)

```

```python
stv_df['Genes']= gene_info_df[gene_info_df['feature_space'] == 'landmark'].index
display(stv_df)
```

```python
survival_coef = pd.read_excel('/home/jing/Phd_project/project_UCD_blca/blca_publication_OUTPUT/blca_publication_OUTPUT_survival/blca_glmnet.xlsx',
              sheet_name='glmnet_result',index_col=0)
survival_coef.columns=['coefs']
survival_coef.dropna(inplace=True)
display(survival_coef)
```

```python
for i in survival_coef.index:
    stv_df.loc[stv_df['Genes']==i,'blca_survival']= survival_coef.loc[i,'coefs']
```

```python
stv_df
```

```python
m_h_genes = pd.read_pickle('/home/jing/Phd_project/project_UCD_blca/blca_OUTPUT/blca_OUTPUT_m_h_convert/m_h_genes.pkl')
display(m_h_genes)
```

```python
stv_df = stv_df.merge(m_h_genes,how='left',left_on='Genes',right_on ='Human_Genes')
```

```python
stv_df
```

```python
stv_df.set_index('Gene',inplace=True)
```

```python
inv_stv = pd.read_pickle(os.path.join(datadir,'stv_lincs_sct_inv_blca_N_M.pkl'))
display(inv_stv)
```

```python
onc_stv = pd.read_pickle(os.path.join(datadir,'stv_lincs_sct_onc_healthy_onc.pkl'))
display(onc_stv)
```

```python
for i in inv_stv.index:
    stv_df.loc[i,'blca_invasiveness'] =inv_stv.loc[i,'n']
for i in onc_stv.index:
    stv_df.loc[i,'blca_oncogenesis'] =onc_stv.loc[i,'n']
```
